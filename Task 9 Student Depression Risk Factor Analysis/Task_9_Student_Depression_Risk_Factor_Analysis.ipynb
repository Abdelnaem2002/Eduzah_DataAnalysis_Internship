```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("/content/Student Depression Dataset.csv")
df.head()


```

```python
df.isna().sum()
```

```python
# Depression distribution
sns.countplot(x="Depression", data=df)
plt.show()
```

```python
# Depression by Gender
sns.countplot(x="Gender", hue="Depression", data=df)
plt.show()
```

```python
# Depression vs Academic Pressure
sns.boxplot(x="Depression", y="Academic Pressure", data=df)
plt.show()
```

```python
# Correlation heatmap
df["Depression_encoded"] = df["Depression"].map({"Yes": 1, "No": 0})
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```
